```python
import pandas as pd
import numpy as np
```

```python
#Loading all data frames into our python notebook

data78 = pd.read_csv('data78.csv', low_memory=False)
data89 = pd.read_csv('data89.csv', low_memory=False)
data910 = pd.read_csv('data910.csv', low_memory=False)
data1011 = pd.read_csv('data1011.csv', low_memory=False)
data1112 = pd.read_csv('data1112.csv', low_memory=False)
data1213 = pd.read_csv('data1213.csv', low_memory=False)
data1314 = pd.read_csv('data1314.csv', low_memory=False)
data1415 = pd.read_csv('data1415.csv', low_memory=False)
data1516 = pd.read_csv('data1516.csv', low_memory=False)
data1617 = pd.read_csv('data1617.csv', low_memory=False)
```

```python
#loading column dictionary

cnames=pd.read_csv('c_name.csv')
```

```python
#taking our column names from the dictionary and converting it to a list

c_list=cnames['column name'].values.tolist()
```

```python
#Extracting all the column names that we require

data1=data78[[c for c in data78.columns if c in c_list]]
data2=data89[[c for c in data89.columns if c in c_list]]
data3=data910[[c for c in data910.columns if c in c_list]]
data4=data1011[[c for c in data1011.columns if c in c_list]]
data5=data1112[[c for c in data1112.columns if c in c_list]]
data6=data1213[[c for c in data1213.columns if c in c_list]]
data7=data1314[[c for c in data1314.columns if c in c_list]]
data8=data1415[[c for c in data1415.columns if c in c_list]]
data9=data1516[[c for c in data1516.columns if c in c_list]]
data10=data1617[[c for c in data1617.columns if c in c_list]]
```

```python
#Exporting our new dataframes

data1.to_csv('data1.csv')
data2.to_csv('data2.csv')
data3.to_csv('data3.csv')
data4.to_csv('data4.csv')
data5.to_csv('data5.csv')
data6.to_csv('data6.csv')
data7.to_csv('data7.csv')
data8.to_csv('data8.csv')
data9.to_csv('data9.csv')
data10.to_csv('data10.csv')
```

```python
# Finding null values in dataframe and storing the column names with null values in a list

df1_na= data1.isna()
count_true=0
count_false=0
list_null_col1= []
for c in c_list:
#     print(c)
    col= list(df1_na[c])
    for r in col:
        if r == True:
            count_true+=1
#             print(count_true)
        else:
            count_false+=1
    if count_true >= count_false:
        list_null_col1.append(c)

# print(list_null_col)
            
```
